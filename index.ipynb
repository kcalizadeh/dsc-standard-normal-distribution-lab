## Load, visualize and give general comments about the dataset

Use pandas for loading and inspecting the data.

```python
# Import libraries
import numpy as np
import seaborn as sns
import pandas as pd 

# Read the yield data as a dataframe
df = pd.read_csv('yield.csv')
df['0'].mean()
```

```python
# Create a plot
sns.distplot(df)
```

## Briefly explain what each value represents in this data set


Each value represents the weight in apples coming from a given tree

## Define the interval bounds which contain 99% of the observations   

> **Hint**: Recall the empirical rule related to $3\sigma$.

```python
# Perform any calculations necessary here
df.std()
interval = [(df['0'].mean() - df['0'].std() * 3), (df['0'].mean() + df['0'].std() * 3)]

std = df['0'].std()
mean = df['0'].mean()
```

```python
# Write your answer here 
interval
```

## Compute and interpret the z-score for a tree yielding 35 pounds of apples

```python
# Calculate z
(35 - mean) / std
```

Interpret the result
this apple is a little over 1 standard deviation below the mean. it is relatively unlikely that an tree would yield this number of apples.

## Suppose a tree has a z-score of 1.85. Interpret this z-score. What is the yield of this tree?

Interpret the z score
this tree has a relatively high yield, almost two standard deviations above the mean

```python
# Calculate yield
output = 1.85 * std + mean
```

```python
# What is the yield ?
output
```

##  Convert each tree’s yield to a z-score so the new variable is the “z-score for weight”

The units are still the apple trees. For the data set of all z-scores:

* What is the shape? 
* The mean? 
* The standard deviation?

```python
# Give your solution
df['standardized'] = (df['0'] - mean) / std
df['standardized'].mean(), df['standardized'].std()
```

```python
# Your observations
# the mean is all but 0, the standard deviation is 1. the distribution is as it was before.
sns.distplot(df['standardized'])
```
